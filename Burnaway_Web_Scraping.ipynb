```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Weds Feb  2 13:20:25 2020
This script obtains web content.

@author: rachelcalder

https://burnaway.org/page/11/?s=wonderroot/
"""
import requests
from bs4 import BeautifulSoup 
import pandas as pd 

df = pd.DataFrame()
# Gathering the page links from search
test_link = []
for page in range(1,12):
      url = f"https://burnaway.org/page/{page}/?s=wonderroot"
      #obtain website content
      response = requests.get(url)
      content = response.content
      #restructure url data for readability
      soup = BeautifulSoup(content, "lxml")

      headers = soup.find_all("h2")
      for header in headers:
        link = header.find('a')
        test_link.append(link.get('href'))

photoCaptions = []
title = []
author = []
date = []
description = []
for page in test_link:
      #obtain website content
      response = requests.get(page)
      content = response.content
      #restructure url data for readability
      soup = BeautifulSoup(content, "lxml")
      article = soup.find_all('article')
      if len(article) >= 1:
        test = article[0].get_text('\n').split('\n')
        #no photo caption
        if test[4] == '·':
          photoCaptions.append('N/A')
          title.append(test[2])
          author.append(test[3])
          date.append(test[5])

          text1 = ''
          for i in test[test.index(' '):]:
            if i.startswith('(function(d, s, id)'):
              break
            if i != '':
              text1 += i + ' '
          description.append(text1)
        #Photo caption is two lines
        elif test[6] == '·':
          photoCaptions.append(test[2] + ' ' + test[3])
          title.append(test[4])
          author.append(test[5])
          date.append(test[7])

          text1 = ''
          for i in test[test.index(' '):]:
            if i.startswith('(function(d, s, id)'):
              break
            if i != '':
              text1 += i + ' '
          description.append(text1)
        #Photo caption has two lines following
        elif test[7] == '·':
          photoCaptions.append(test[2])
          title.append(test[5])
          author.append(test[6])
          date.append(test[8])

          text1 = ''
          for i in test[test.index(' '):]:
            if i.startswith('(function(d, s, id)'):
              break
            if i != '':
              text1 += i + ' '
          description.append(text1)
        #Photo caption is one line
        else:
          photoCaptions.append(test[2])
          title.append(test[3])
          author.append(test[4])
          date.append(test[6])

          text1 = ''
          for i in test[test.index(' '):]:
            if i.startswith('(function(d, s, id)'):
              break
            if i != '':
              text1 += i + ' '
          description.append(text1)
      else:
        photoCaptions.append(' ')
        title.append(' ')
        author.append(' ')
        date.append(' ')
        description.append(' ')

df['tite'] = title
df['link'] = test_link
df['author'] = author
df['date'] = date
df['photo caption'] = photoCaptions
df['description'] = description

df.to_csv("wonderrootArtsATL.csv")
```
